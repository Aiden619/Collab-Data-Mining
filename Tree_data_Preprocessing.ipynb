```python
tree_census['health'].fillna('good',inplace = True)
tree_census['spc_latin'].fillna('No observation',inplace = True)
tree_census['sidewalk'].fillna('NoDamage',inplace = True)
tree_census.fillna('None',inplace = True)
```

```python
tree = tree_census.loc[:,['tree_dbh', 'stump_diam']]
# copy the data
df_max_scaled = tree.copy()
# apply normalization techniques
for column in df_max_scaled.columns:
  df_max_scaled[column] = df_max_scaled[column] / df_max_scaled[column].abs().max()
# view normalized data
display(df_max_scaled)
```

```python
# Creating the data frame
details = {
'col1': [1, 3, 5, 7, 9],
'col2': [7, 4, 35, 14, 56]
}
# creating a Dataframe object
df = pd.DataFrame(details)
# Z-Score using pandas
df['col1'] = (df['col1'] - df['col1'].mean()) / df['col1'].std()
df
```

```python
# Importing LabelEncoder from Sklearn library from preprocessing Module.
from sklearn.preprocessing import LabelEncoder
df = pd.DataFrame({
    'fruit':  ['apple','orange','apple','orange'],
    'color':  ['red','orange','green','green'],
    'weight': [5,6,3,4]
})
# Creating a instance of label Encoder.
le = LabelEncoder()
# Using .fit_transform function to fit label encoder and return encoded
label = le.fit_transform(df['fruit'])
# printing label
label
```
